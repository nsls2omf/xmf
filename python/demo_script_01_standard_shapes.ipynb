# Demo script_01 standard shapes

```python
import numpy as np
import xmf
```

```python
# 1. Lateral coordinates
x_range = 200e-3
y_range = 20e-3
x_num = 201
y_num = 21
x1d = np.linspace(-x_range/2, x_range/2, x_num)
y1d = np.linspace(-y_range/2, y_range/2, y_num)
x2d, y2d = np.meshgrid(x1d, y1d)

abs_p = 30
abs_q = 0.3
theta = 30e-3
```

```python
# 2.1. Convex Ellipsoid (CVXE)

z2d_quad_sln, z2d_expression = xmf.standard_convex_ellipsoid_height(x2d, y2d, abs_p, abs_q, theta, return_z2d_expression_as_extra=True)
xmf.fig_show_2d_map(x2d, y2d, z2d_quad_sln, z2d_expression, 'Convex Ellipsoid')
```

```python
#  2.2. Concave Ellipsoid (CCVE)

z2d_quad_sln, z2d_expression = xmf.standard_concave_ellipsoid_height(x2d, y2d, abs_p, abs_q, theta, return_z2d_expression_as_extra=True)
xmf.fig_show_2d_map(x2d, y2d, z2d_quad_sln, z2d_expression, 'Concave Ellipsoid')
```

```python
# 3.1. Convex Elliptic Cylinder (CVXEC)

z2d_quad_sln, z2d_expression = xmf.standard_convex_elliptic_cylinder_height(x2d, abs_p, abs_q, theta, return_z_expression_as_extra=True)
xmf.fig_show_2d_map(x2d, y2d, z2d_quad_sln, z2d_expression, 'Convex Elliptic Cylinder')

z1d_quad_sln, z1d_expression = xmf.standard_convex_elliptic_cylinder_height(x1d, abs_p, abs_q, theta, return_z_expression_as_extra=True)
xmf.fig_show_1d_height(x1d, z1d_quad_sln, z1d_expression, 'Convex Elliptic Cylinder')

sx1d = xmf.standard_convex_elliptic_cylinder_xslope(x1d, abs_p, abs_q, theta)
xmf.fig_show_1d_slope(x1d, sx1d, 'Convex Elliptic Cylinder')

```

```python
# 3.2. Concave Elliptic Cylinder (CCVEC)

z2d_quad_sln, z2d_expression = xmf.standard_concave_elliptic_cylinder_height(x2d, abs_p, abs_q, theta, return_z_expression_as_extra=True)
xmf.fig_show_2d_map(x2d, y2d, z2d_quad_sln, z2d_expression, 'Concave Elliptic Cylinder')

z1d_quad_sln, z1d_expression = xmf.standard_concave_elliptic_cylinder_height(x1d, abs_p, abs_q, theta, return_z_expression_as_extra=True)
xmf.fig_show_1d_height(x1d, z1d_quad_sln, z1d_expression, 'Concave Elliptic Cylinder')

sx1d = xmf.standard_concave_elliptic_cylinder_xslope(x1d, abs_p, abs_q, theta)
xmf.fig_show_1d_slope(x1d, sx1d, 'Concave Elliptic Cylinder')
```

```python
# 6.1. Convex Hyperboloid (CVXH)

z2d_quad_sln, z2d_expression = xmf.standard_convex_hyperboloid_height(x2d, y2d, abs_p, abs_q, theta, return_z2d_expression_as_extra=True)
xmf.fig_show_2d_map(x2d, y2d, z2d_quad_sln, z2d_expression, 'Convex Hyperboloid')
```
