# Demo script_02 generation of shapes

```python
import numpy as np
import xmf
```

```python
# 1. Lateral coordinates
x_range = 200e-3
y_range = 20e-3
x_num = 201
y_num = 21
x1d = np.linspace(-x_range/2, x_range/2, x_num)
y1d = np.linspace(-y_range/2, y_range/2, y_num)
[x2d, y2d] = np.meshgrid(x1d, y1d)

abs_p = 30
abs_q = 0.3
theta = 30e-3

x_i = 0*-1e-3
y_i = 0*-2e-4
z_i = 0*3e-7
alpha = 0*2e-6
beta = 0*1e-5
gamma = 1.0e-3
```

```python
#  2.1. Convex Ellipsoid (CVXE)

z2d_standard = xmf.standard_convex_ellipsoid_height(x2d, y2d, abs_p, abs_q, theta)
z2d = xmf.generate_2d_curved_surface_height(xmf.standard_convex_ellipsoid_height, x2d, y2d, abs_p, abs_q, theta, x_i, y_i, z_i, alpha, beta, gamma)
xmf.fig_show_2d_maps(x2d, y2d, z2d, z2d_standard, 'Convex Ellipsoid')

```

```python
#  2.2. Concave Ellipsoid (CCVE)

z2d_standard = xmf.standard_concave_ellipsoid_height(x2d, y2d, abs_p, abs_q, theta)
z2d = xmf.generate_2d_curved_surface_height(xmf.standard_concave_ellipsoid_height, x2d, y2d, abs_p, abs_q, theta, x_i, y_i, z_i, alpha, beta, gamma)
xmf.fig_show_2d_maps(x2d, y2d, z2d, z2d_standard, 'Concave Ellipsoid')
```

```python
# 3.1. Convex Elliptic Cylinder (CVXEC)

z2d_standard = xmf.standard_convex_elliptic_cylinder_height(x2d, abs_p, abs_q, theta)
z2d = xmf.generate_2d_cylinder_height(xmf.standard_convex_elliptic_cylinder_height, x2d, y2d, abs_p, abs_q, theta, x_i, z_i, alpha, beta, gamma)
xmf.fig_show_2d_maps(x1d, y1d, z2d, z2d_standard, 'Convex Elliptic Cylinder')

z1d_standard = xmf.standard_convex_elliptic_cylinder_height(x1d, abs_p, abs_q, theta)
z1d = xmf.generate_1d_height(xmf.standard_convex_elliptic_cylinder_height, x1d, abs_p, abs_q, theta, x_i, z_i, beta)
xmf.fig_compare_1d_height(x1d, z1d, z1d_standard, 'Convex Elliptic Cylinder')

sx1d_standard = xmf.standard_convex_elliptic_cylinder_xslope(x1d, abs_p, abs_q, theta)
sx1d = xmf.generate_1d_slope(xmf.standard_convex_elliptic_cylinder_xslope, x1d, abs_p, abs_q, theta, x_i, beta)
xmf.fig_compare_1d_slope(x1d, sx1d, sx1d_standard, 'Convex Elliptic Cylinder')
```

```python
# 3.2. Concave Elliptic Cylinder (CCEC)

z2d_standard = xmf.standard_concave_elliptic_cylinder_height(x2d, abs_p, abs_q, theta)
z2d = xmf.generate_2d_cylinder_height(xmf.standard_concave_elliptic_cylinder_height, x2d, y2d, abs_p, abs_q, theta, x_i, z_i, alpha, beta, gamma)
xmf.fig_show_2d_maps(x1d, y1d, z2d, z2d_standard, 'Concave Elliptic Cylinder')

z1d_standard = xmf.standard_concave_elliptic_cylinder_height(x1d, abs_p, abs_q, theta)
z1d = xmf.generate_1d_height(xmf.standard_concave_elliptic_cylinder_height, x1d, abs_p, abs_q, theta, x_i, z_i, beta)
xmf.fig_compare_1d_height(x1d, z1d, z1d_standard, 'Concave Elliptic Cylinder')

sx1d_standard = xmf.standard_concave_elliptic_cylinder_xslope(x1d, abs_p, abs_q, theta)
sx1d = xmf.generate_1d_slope(xmf.standard_concave_elliptic_cylinder_xslope, x1d, abs_p, abs_q, theta, x_i, beta)
xmf.fig_compare_1d_slope(x1d, sx1d, sx1d_standard, 'Concave Elliptic Cylinder')
```
