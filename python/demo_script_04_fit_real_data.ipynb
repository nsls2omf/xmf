# Demo script_04 fit real data

```python
import os
import numpy as np
from scipy.io import loadmat
import xmf
```

```python
map = loadmat(os.path.join('..', 'real_data', 'sample_03_concave_ellipsoid_height_map.mat'))
```

```python
p = map['params_target'][0][0][0][0][0]
q = map['params_target'][0][0][1][0][0]
theta = map['params_target'][0][0][2][0][0]
x_i = map['params_target'][0][0][3][0][0]
y_i = map['params_target'][0][0][4][0][0]
z_i = map['params_target'][0][0][5][0][0]
alpha = map['params_target'][0][0][6][0][0]
beta = map['params_target'][0][0][7][0][0]
gamma = map['params_target'][0][0][8][0][0]

x2d = map['x2d']
y2d = map['y2d']
z2d_measured = map['z2d']
print(p, q, theta, x_i, y_i, z_i, alpha, beta, gamma)

target_params_dict = {
    'p': p,
    'q': q,
    'theta': theta,
    'x_i': x_i,
    'y_i': y_i,
    'z_i': z_i,
    'alpha': alpha,
    'beta': beta,
    'gamma': gamma
}

params_input_dict = {
    'p': p,
    'q': q,
    'theta': theta,    
    'y_i': 0.0,
}

is_opt_dict = {
    'p': False,
    'q': False,
    'theta': False,
    'y_i': False,
}
```

```python
z2d_res, z2d_fit, opt_params_dict, is_optimized_dict = xmf.fit_concave_ellipsoid_height(x2d, y2d, z2d_measured, params_input_dict, is_opt_dict, is_lmfit=True) 
xmf.fig_show_2d_fitting_maps(x2d, y2d, z2d_measured, z2d_fit, z2d_res, target_params_dict, opt_params_dict, is_optimized_dict,'Concave Ellipsoid')
opt_params_dict 
```
