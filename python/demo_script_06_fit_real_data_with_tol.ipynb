# Demo script_06 fit real data with tolerance option

```python
import os
from scipy.io import loadmat
import xmf
```

```python
concave_ellipse_slope = loadmat(os.path.join('..', 'real_data', 'sample_01_concave_elliptic_cylinder_slope.mat'))

p = concave_ellipse_slope['params_target'][0][0][0][0][0]
q = concave_ellipse_slope['params_target'][0][0][1][0][0]
theta = concave_ellipse_slope['params_target'][0][0][2][0][0]
x_i = concave_ellipse_slope['params_target'][0][0][3][0][0]
beta = concave_ellipse_slope['params_target'][0][0][4][0][0]

x1d = concave_ellipse_slope['x1d'].flatten()
sx1d_measured = concave_ellipse_slope['sx1d'].flatten()

target_params_dict = {
    'p': p,
    'q': q,
    'theta': theta,
    'x_i': x_i,
    'beta': beta,
}

params_input_dict = {
    'p': p,
    'q': q,
    'theta': theta
}

tol_dict = {
    'p': 0,
    'q': 0,
    'theta': 0
}

sx1d_res, sx1d_fit, opt_params_dict, opt_params_ci_dict, _ = xmf.fit_concave_ellipse_slope(x1d, sx1d_measured, params_input_dict, tol_dict) 
xmf.fig_show_1d_fitting_slope(x1d, sx1d_measured, sx1d_fit, sx1d_res, target_params_dict, opt_params_dict, opt_params_ci_dict, 'Concave Ellipse Slope') 
```
